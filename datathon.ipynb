# this function is used to convert 'l' and 'o' into 1 and 0 of the column s52

```python
def s52_convert(text):
    if text == '1' or text == '0':
        return text
    else:
        if text == 'l':
            return '1'
        else:
            return '0'
```

## convertion starts

```python
data['s52mod'] = data['s52'].apply(s52_convert)
```

# by calling the function s52 is modified to s52mod

```python
def op_remover(text):
    if 'A' in text:
        return 'A'
    elif 'B' in text:
        return 'B'
    elif 'C' in text:
        return 'C'
    elif 'D' in text:
        return 'D'
```

```python
data['s70mod'] = data['s70'].apply(op_remover)
```
